```python
#  transform slide placeholder names
def transform_placeholder_names(name):
    if "Segnaposto titolo" in name:
       return "slide_header"
    if "Text Placeholder" in name:
        return "text_placeholder"
    if "testo" in name:
        return "text_placeholder"
    if "numero diapositiva" in name:
        return "slide_number"
    if  "Titolo" in name:
        return "presentation_title"
    
```

```python
new_ppt=Presentation("recordatti-master.pptx")
from pptx.util import Inches, Pt
from pptx.enum.shapes import MSO_SHAPE
from pptx.dml.color import RGBColor
import re
from pptx.enum.text import PP_ALIGN
```

```python
def remove_numbers(text):
    return re.sub(r'^\d+\.\s*', '', text)
```

```python
def parse_text_data(text):
    slides = []
    slide = {}
    lines = text.strip().split('\n')
    
    for line in lines:
        if line.startswith("#Title:"):
            title = line.replace("#Title:", "").strip()
            slide['title'] = title
        elif line.startswith("#Slide:"):
            if slide:
                slides.append(slide)
                slide = {}
            slide_number = line.replace("#Slide:", "").strip()
            slide['slide_number'] = slide_number
        elif line.startswith("#Header:") or line.startswith("#Headers:"):
            header = line.replace("#Header:", "").replace("#Headers:", "").strip()
            slide['header'] = header
        elif line.startswith("#Content:"):
            content = line.replace("#Content:", "").strip()
            slide['content'] = content
        else:
            if 'content' in slide:
                slide['content'] += '\n' + line.strip()
    
    if slide:
        slides.append(slide)
    
    return slides


```

```python
# Function to create PowerPoint presentation
def create_presentation(slides, output_file):
   
    #  Title Slide
    ref_slide=slide_selection["intro"]
    # title_placeholder_index=ref_slide['placeholders'].keys()[0]
    # print(list(ref_slide['placeholders'].k())[0])
    title_placeholder_index=list(ref_slide['placeholders'].values())[0]
    #  For title Slide we have only one placeholder so choose the first one itself.
    slide = new_ppt.slides.add_slide(ref_slide['slide_obj'])
    title = slide.placeholders[title_placeholder_index]
    title.text = slides[0]['title']
    
    # ref_slide=get_slide(1,0)
    ref_slide=slide_selection['table_of_content']
    slide = new_ppt.slides.add_slide(ref_slide['slide_obj'])
    title_placeholder_index=ref_slide['placeholders']['slide_header']
    title = slide.placeholders[title_placeholder_index]
    title.text = slides[1]['header']
    
    lines = slides[1]['content'].split('\n')
    bullet_color = RGBColor(255, 0, 0)
    left = Inches(0.5)
    top = Inches(1)
    width = Inches(8.5)
    height = Inches(5.5)
    textbox = slide.shapes.add_textbox(left, top, width, height)
    text_frame = textbox.text_frame
    print(lines)
    for idx, line in enumerate([remove_numbers(line) for line in lines]):
    
        if(line.strip()!=''):
        
            p = text_frame.add_paragraph()
            p.level = 0  # Bullet point level
            p.font.size = Pt(18)
            p.font.color.rgb = RGBColor(0, 0, 0)
            p.alignment = PP_ALIGN.LEFT
            
            # Add a custom bullet at the beginning
            run = p.add_run()
            run.text = '\u25BA'  # Unicode character for a triangle
            run.font.color.rgb = bullet_color
            run.font.size = Pt(18)
            run.font.bold = True
            
            # Add the text after the bullet
            print(idx)
            run = p.add_run()
            run.text = ' ' + line.strip()
            run.font.color.rgb = RGBColor(0, 0, 0)
            run.font.size = Pt(18)
   
     
    # Add other slides
    for slide_data in slides[2:]:
        # ref_slide =get_slide(1,1)
        ref_slide=slide_selection['content_slide']
        print(ref_slide)
        
        slide = new_ppt.slides.add_slide(ref_slide['slide_obj'])
        title_placeholder_index=ref_slide['placeholders']['slide_header']
        # content = slide.placeholders[1]
        content_placeholder_index=ref_slide['placeholders']['text_placeholder']
        title = slide.placeholders[title_placeholder_index]
                        
        content=slide.placeholders[content_placeholder_index]
        # slide_number=slide.placeholders[slide_number_placeholder_index]
        title.text = slide_data['header']
        if "table of contents" in slide_data['header'].lower():
    
            content.text = ""  # Clear the content placeholder text
            lines = slide_data['content'].split('\n')
            bullet_color = RGBColor(255, 0, 0)
            left = Inches(0.5)
            top = Inches(1)
            width = Inches(8.5)
            height = Inches(5.5)
            textbox = slide.shapes.add_textbox(left, top, width, height)
            text_frame = textbox.text_frame
            print(lines)
            for idx, line in enumerate([remove_numbers(line) for line in lines]):

                if(line.strip()!=''):
                
                     p = text_frame.add_paragraph()
                     p.level = 0  # Bullet point level
                     p.font.size = Pt(18)
                     p.font.color.rgb = RGBColor(0, 0, 0)
                     p.alignment = PP_ALIGN.LEFT
                   
                    
                        # Add a custom bullet at the beginning
                     run = p.add_run()
                     run.text = '\u25BA'  # Unicode character for a triangle
                     # run.space_before = Pt(28)  
                     run.font.color.rgb = bullet_color
                     run.font.size = Pt(18)
                     run.font.bold = True
                
                    # Add the text after the bullet
                     print(idx)
                     run = p.add_run()
                     run.text = ' ' + line.strip()
                     run.font.color.rgb = RGBColor(0, 0, 0)
                     run.font.size = Pt(18)
                                
               
        else:
            content.text = slide_data['content']

        # slide_number.text=slide_data['slide_number']
    # Add end Slide

    ref_slide =get_slide(0,7)
    print(ref_slide)
        
    slide = new_ppt.slides.add_slide(ref_slide['slide_obj'])
    
    
    # Save the presentation
    print(output_file)
    # desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")

# Define the filename
    filename = "gen.pptx"

# Full path to save the file
    # full_path = os.path.join(desktop_path, filename)
    
    new_ppt.save(filename)

    print(f"PowerPoint presentation created successfully and saved as {output_file}.")

# Function to read the text file and generate the presentation
```

```python
async def create_ppt_text(input, messages):
        print(input)
        # Take only the latest 2 conversationsCan
        messages = messages[-4:]
        print(messages)
        history = """"""
        for i in range(len(messages)):
            message_content = messages[i].get('content')
            message_content_cleaned = re.sub(r'<a[^>]*>(.*?)</a>', r'\1', message_content)

            history += f"""{messages[i].get('role')}: {message_content_cleaned}"""
            history += "\n"

        message = f"""The user wants a presentation about "{input}"

History:
{history}
"""

        response = openai_client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": get_system_message()},
                {"role": "user", "content": message}
            ],
            temperature=0.5,
        )

        chat_coroutine_result = await response
        chat_resp = chat_coroutine_result.model_dump()
        print(chat_resp)
        return chat_resp["choices"][0]["message"]["content"]

   
```
